Anteriormente, foram feitas ànalises dos dados dentro do DataSet por meio da Ferramenta Microsoft Excel.
Através desta análise, foram extraídos os seguintes insights sobre este DataSet:


- O dataset conta com mais de 57 mil registros de acidentes. Dentre estes
registros (As respectivas variáveis estão listadas Acima no comando data.info() )
- Em meio a estes registros, existem dados incompletos e/ou que
necessitem de uma “limpeza”, ou seja, estão incompletos. Dificultando
assim o entendimento e melhor coleta de informações.
Por isso, os registros onde o Agente Causador, CBO da vítima e CID
estão (ambos) como nulos, foram descartados. - Comando será realizado e sinalizado posteriormente
- Foi constatado que haviam também registros

E desses 254 registros, 05 se tratavam de Doenças, onde

Natureza da Lesão é classificada como doenças! Este tipo de Registro
não tem muita valia para proposta do SafeZone, que busca apenas
registros de acidentes dentro de empresas e/ou áreas de construção.,
portanto estes registros não serão considerados.
- Foram desconsiderados 2972 registros onde não se especificou o
agente causador e/ou também o CBO da vítima, inviabilizando a
identificação do dado (se faz parte ou não da proposta do SafeZone)

Limpeza dos dados Incompletos (onde AgenteCausador, CBO da Vitima e CID estão ambos como nulos ( {ñ class} )


```python
# Verificando inicialmente quantos registros possuem "{ñ class}" nas colunas especificadas
print(data[(data['Agente_Causador_Acidente'] == '{ñ class}') &
           (data['CBO'] == '{ñ class}') &
           (data['CID-10'] == '{ñ class}')].shape[0])
```

```python
# Removendo registros onde 'Agente_Causador_Acidente', 'CBO' e 'CID-10' são simultaneamente '{ñ class}'
data = data[~((data['Agente_Causador_Acidente'] == '{ñ class}') &
              (data['CBO'] == '{ñ class}') &
              (data['CID-10'] == '{ñ class}'))]
```

```python
print(data.shape)
print(data[(data['Agente_Causador_Acidente'] == '{ñ class}') &
           (data['CBO'] == '{ñ class}') &
           (data['CID-10'] == '{ñ class}')])
```

Remoção dos dados onde a Natureza da lesão é classificada como "doença"

```python
# Removendo registros onde 'Natureza_Lesao' contém a palavra "doenca"
data = data[~data['Natureza_Lesao'].str.contains('doenca', case=False, na=False)]
```

```python
# Verificando se os registros foram removidos
print(data.shape)
print(data['Natureza_Lesao'].unique())
```

Exclusão dos dados {ñ class} contidos em Agente_Causador_Acidente e/ou CBO


```python
# Removendo registros onde 'Agente_Causador_Acidente' é '{ñ class}'
data = data[data['Agente_Causador_Acidente'] != '{ñ class}']
```

```python
# Verificando se os registros foram removidos
print(data.shape)
print(data['Agente_Causador_Acidente'].unique())
```

```python
# Removendo registros onde 'CBO' é '{ñ class}'
data = data[data['CBO'] != '{ñ class}']

# Verificando se os registros foram removidos
print(data.shape)
print(data['CBO'].unique())
```

Exclusão dos demais dados fora da proposta do SafeZone

```python
# Removendo registros onde 'Agente_Causador_Acidente' é '{Agente do Acidente Inexistente }'
data = data[data['Agente_Causador_Acidente'] != 'Agente do Acidente Inexistente ']

# Verificando se os registros foram removidos
print(data.shape)
print(data['Agente_Causador_Acidente'].unique())
```

```python
# Definindo as condições de filtro
condition = ((data['Agente_Causador_Acidente'] != 'Coque') &
             (~data['CID-10'].isin(['W86.4 Rua e Estrada', 'W85.9 Local Ne', 'W87.7 Fazenda'])))

# Aplicando o filtro
data = data[condition]

# Verificando se os registros foram removidos
print(data.shape)
print(data[['Agente_Causador_Acidente', 'CID-10']].head())
```

```python
# Lista de valores a serem excluídos
agentes_a_excluir = [
    'Aeronave',
    'Agente Infeccioso ou Parasitario - Inclui Bac',
    'Animal Vivo',
    'Ataque de Ser Vivo com Peconha',
    'Ataque de Ser Vivo com Transmissao de Doenca',
    'Ataque de Ser Vivo por Mordedura, Picada, Chi',
    'Ataque de Ser Vivo, Nic',
    'Bicicleta',
    'Choque',
    'Couro Cru ou Curtido - Produto Animal',
    'Esforco Excessivo ao Empurrar ou Puxar Objeto',
    'Esforco Excessivo ao Erguer Objeto',
    'Esforco Excessivo ao Manejar, Sacudir ou Arre',
    'Esforco Excessivo, Nic',
    'Ferro de Passar- Ferramenta Portatil com Forc',
    'Fruta e Derivados',
    'Leite e Derivados - Inclusive de Origem Anima',
    'Metal - Inclui Liga Ferrosa e nao Ferrosa, Tu',
    'Motocicleta, Motoneta',
    'Osso - Produto Animal',
    'Pele, Crina, Pelo, La (Em Bruto) - Produto An',
    'Piso de Veiculo - Superficie Utilizada para S',
    'Produto Alimenticio - Inclusive de Origem Ani',
    'Produto Animal, Nic',
    'Puncao, Ponteiro, Vazador, Talhadeira- Ferram',
    'Ser Vivo, Nic',
    'Vegetal - Planta, Arvore, em Estado Natural,',
    'Veiculo Aquatico',
    'Veiculo de Terraplenagem',
    'Veiculo de Tracao Animal',
    'Veiculo Deslisante',
    'Veiculo Rodoviario Motorizado',
    'Veiculo Sobre Trilho',
    'Veiculo, Nic',
    'Vestuario, Nic'
]

# Removendo registros onde 'Agente_Causador_Acidente' está na lista de valores a excluir
data = data[~data['Agente_Causador_Acidente'].isin(agentes_a_excluir)]

# Verificando se os registros foram removidos
print(data.shape)
print(data['Agente_Causador_Acidente'].unique())
```

```python
# Definindo as condições de filtro
condition = ((data['Agente_Causador_Acidente'] != 'Coque') &
             (~data['CID-10'].isin(['W86.4 Rua e Estrada', 'W85.9 Local Ne', 'W87.7 Fazenda'])))

# Aplicando o filtro
data = data[condition]

# Verificando se os registros foram removidos
print(data.shape)
print(data[['Agente_Causador_Acidente', 'CID-10']].head())
```

Em suma: após o tratamento dos dados, foram mantidos 37.555 registros de
ocorrências, que fazem parte da proposta do SafeZone.
Ou seja, foram descartados 19.944 registros que não fazem parte do
escopo principal do projeto, fazendo com que a busca por informações
seja mais rápida e também assertiva.

```python
import matplotlib.pyplot as plt

# Agrupando os dados pelo 'Agente_Causador_Acidente' e contando o número de ocorrências de cada acidente
top_10_acidentes = data['Agente_Causador_Acidente'].value_counts().nlargest(10)

import matplotlib.pyplot as plt

# Agrupando os dados pelo 'Agente_Causador_Acidente' e contando o número de ocorrências de cada acidente
top_10_acidentes = data['Agente_Causador_Acidente'].value_counts().nlargest(10)

# Criando um gráfico de barras para os 10 acidentes com maior ocorrência
plt.figure(figsize=(10, 6))
top_10_acidentes.plot(kind='bar', color='skyblue')
plt.title('Top 10 Acidentes com Maior Ocorrência')
plt.xlabel('Agente Causador do Acidente')
plt.ylabel('Número de Ocorrências')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
